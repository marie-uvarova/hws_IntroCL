````python
import pyconll

text = pyconll.load_from_file('myfile.conllu')
disl = 0
disc = 0

for elem in text:
    sent = ''
    DISC = []
    DISL = []
    for token in elem:
        if token.upos != 'PUNCT' and token.id != 1:
            sent += ' '
            sent += token.form
        elif token.form == r'[,.\]':
            sent += token.form        
            sent += ' '
        else:
            sent += token.form

        if token.deprel == 'discourse':
            disc += 1
            DISC.append(token.lemma)
        elif token.deprel == 'dislocated':
            disl += 1
            DISL.append(token.lemma)
            
    if disc > 0:
        for elem in DISC:
            print(elem)
            print(sent)
    if disl > 0:
        for elem in DISL:
            print(elem)
            print(sent)
if disl == 0:
    print('There are no tokens with the label "dislocated"')
elif disc == 0:
    print('There are no tokens with the label "discourse"')
````

Возьмите любой достаточно длинный (лучше новостной) текст. Любым известным инструментом извлеките именованные сущности из этого текста и выведите их списком по категориям (т.е. персоны вместе, локации вместе, организации вместе). 

````python
import spacy
 
nlp = spacy.load('en_core_web_sm')
 
with open('news.txt', 'r', encoding='utf-8') as f:
    text = f.read()
    doc = nlp(text)

NAMES = []
LOC = []
ORG = []

for ent in doc.ents:
    if ent.label_ == 'PERSON':
        NAMES.append(ent.text)
    elif ent.label_ == 'GPE':
        LOC.append(ent.text)
    elif ent.label_ == 'ORG':
        ORG.append(ent.text)

print('Names: ')
for elem in NAMES:
    print(elem, end=', ')
print('\nLocations: ')
for elem in LOC:
    print(elem, end=', ')
print('\nOrganisations: ')
for elem in ORG:
    print(elem, end=', ')
#почему-то много где распознает трампа как организацию...
````
